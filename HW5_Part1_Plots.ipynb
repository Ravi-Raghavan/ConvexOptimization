# Generate Plots for 9.10

```python
import numpy as np
import matplotlib.pyplot as plt

#9.10(a) function
def f1(x):
    return np.log(np.exp(x) + np.exp(-1 * x))

#x values
x = np.linspace(-10, 10, 1000)  

#y1 = f1(x)
y1 = f1(x)

plt.plot(x, y1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('9.10 (a) Function Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#9.10(a) function derivative
def f1_derivative(x):
    return (np.exp(x) - np.exp(-1 * x)) / (np.exp(x) + np.exp(-1 * x))

#x values
x = np.linspace(-10, 10, 1000)  

#y1_derivative = f1_derivative(x)
y1_derivative = f1_derivative(x)

plt.plot(x, y1_derivative)
plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.title('9.10 (a) Function Derivative Plot')
plt.show()
```

```python
#9.10(b) function
def f2(x):
    return (-1 * np.log(x)) + x

#x values
x = np.linspace(1e-6, 10, 1000)  

#y2 = f2(x)
y2 = f2(x)

plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('9.10 (b) Function Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#9.10(b) function derivative
def f2_derivative(x):
    return (x - 1) / (x)

#x values
x_neg = np.linspace(-10, -1e-1, 1000)
x_pos = np.linspace(1e-1, 10, 1000)

#y2_derivative = f2_derivative(x)
y2_derivative_neg = f2_derivative(x_neg)
y2_derivative_pos = f2_derivative(x_pos)

plt.plot(x_neg, y2_derivative_neg, label = 'derivative, x < 0')
plt.plot(x_pos, y2_derivative_pos, label = 'derivative, x > 0')

plt.xlabel('x')
plt.ylabel('f\'(x)')
plt.title('9.10 (b) Function Derivative Plot')
plt.legend()
plt.show()
```
